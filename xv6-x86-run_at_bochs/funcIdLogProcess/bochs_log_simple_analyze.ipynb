## bochs日志按函数id分组

```python
dfLog=pandas.read_csv('/crk/bochs/xv6-x86-run_at_bochs/funcIdLogProcess/bochsout_clean.log')
```

```python
dfLogG = dfLog.groupby('abs_location_id').size().reset_index()
dfLogG.columns = ['函数id', '函数调用次数']

dfLogG=dfLogG.sort_values(by='函数调用次数', ascending=False)
```

```python
dfLogG.head(3)
```

```python
dfLogG['文件名_函数名']=dfLogG.函数id.apply(lambda fId: f"{fId2PN[fId]['filePath'] } : {fId2PN[fId]['funcName'] }" )
```
